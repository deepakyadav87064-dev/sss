Here's a simple Python attendance system. It uses a dictionary to store student attendance status. You can run the functions to add students, mark attendance, and view records.

```python
# A dictionary to store attendance. Key: student name, Value: list of dates attended
attendance_records = {}

def add_student(name):
    if name not in attendance_records:
        attendance_records[name] = []
        print(f"Student '{name}' added.")
    else:
        print(f"Student '{name}' already exists.")

def mark_attendance(student_name, date, present=True):
    if student_name in attendance_records:
        if present:
            if date not in attendance_records[student_name]:
                attendance_records[student_name].append(date)
                print(f"'{student_name}' marked present on {date}.")
            else:
                print(f"'{student_name}' already marked present on {date}.")
        else:
            # For simplicity, we'll just note absence, not store it directly in this list
            # A more robust system would track all days and mark P/A/L
            print(f"'{student_name}' marked absent on {date}.")
    else:
        print(f"Student '{student_name}' not found. Please add the student first.")

def view_attendance(student_name=None):
    if not attendance_records:
        print("No attendance records yet.")
        return

    if student_name:
        if student_name in attendance_records:
            print(f"\nAttendance for {student_name}:")
            if attendance_records[student_name]:
                for date in attendance_records[student_name]:
                    print(f"  - {date}: Present")
            else:
                print("  No attendance recorded for this student yet.")
        else:
            print(f"Student '{student_name}' not found.")
    else:
        print("\n--- All Attendance Records ---")
        for name, dates in attendance_records.items():
            print(f"Student: {name}")
            if dates:
                for date in dates:
                    print(f"  - {date}: Present")
            else:
                print("  No attendance recorded yet.")
        print("----------------------------")

# --- Example Usage ---
print("--- Demonstrating Attendance System ---")
add_student("Alice")
add_student("Bob")
add_student("Alice") # Try adding again

mark_attendance("Alice", "2023-10-26")
mark_attendance("Bob", "2023-10-26")
mark_attendance("Alice", "2023-10-27")
mark_attendance("Charlie", "2023-10-26", present=True) # Student not found
mark_attendance("Bob", "2023-10-26") # Already marked

view_attendance("Alice")
view_attendance("Bob")
view_attendance("Charlie") # Student not found
view_attendance() # View all

# You can clear records or reset for new term
# attendance_records.clear()
```

```python
import pandas as pd

def save_attendance_to_csv(records, filename='attendance_records.csv'):
    # Convert the dictionary to a list of records suitable for DataFrame
    data = []
    for student_name, dates_present in records.items():
        if dates_present:
            for date in dates_present:
                data.append({'Student Name': student_name, 'Date': date, 'Status': 'Present'})
        else:
            # Include students with no attendance recorded yet
            data.append({'Student Name': student_name, 'Date': None, 'Status': 'No Record'})

    if not data:
        print("No attendance records to save.")
        return

    df = pd.DataFrame(data)

    # Sort by student name and date for better readability
    df = df.sort_values(by=['Student Name', 'Date']).reset_index(drop=True)

    df.to_csv(filename, index=False)
    print(f"Attendance records saved to '{filename}'.")

# Call the function to save the current attendance_records
save_attendance_to_csv(attendance_records)
```

For longer text generations, you can stream the response. This displays the output token by token as it's generated, rather than waiting for the entire response to complete. This provides a more interactive and responsive experience. To enable this, simply set stream=True.

```python
# @title Simple streaming example
from google.colab import ai

stream = ai.generate_text("Tell me a short story.", stream=True)
for text in stream:
  print(text, end='')
```

```python
#@title Text formatting setup
#code is not necessary for colab.ai, but is useful in fomatting text chunks
import sys

class LineWrapper:
    def __init__(self, max_length=80):
        self.max_length = max_length
        self.current_line_length = 0

    def print(self, text_chunk):
        i = 0
        n = len(text_chunk)
        while i < n:
            start_index = i
            while i < n and text_chunk[i] not in ' \n': # Find end of word
                i += 1
            current_word = text_chunk[start_index:i]

            delimiter = ""
            if i < n: # If not end of chunk, we found a delimiter
                delimiter = text_chunk[i]
                i += 1 # Consume delimiter

            if current_word:
                needs_leading_space = (self.current_line_length > 0)

                # Case 1: Word itself is too long for a line (must be broken)
                if len(current_word) > self.max_length:
                    if needs_leading_space: # Newline if current line has content
                        sys.stdout.write('\n')
                        self.current_line_length = 0
                    for char_val in current_word: # Break the long word
                        if self.current_line_length >= self.max_length:
                            sys.stdout.write('\n')
                            self.current_line_length = 0
                        sys.stdout.write(char_val)
                        self.current_line_length += 1
                # Case 2: Word doesn't fit on current line (print on new line)
                elif self.current_line_length + (1 if needs_leading_space else 0) + len(current_word) > self.max_length:
                    sys.stdout.write('\n')
                    sys.stdout.write(current_word)
                    self.current_line_length = len(current_word)
                # Case 3: Word fits on current line
                else:
                    if needs_leading_space:
                        # Define punctuation that should not have a leading space
                        # when they form an entire "word" (token) following another word.
                        no_leading_space_punctuation = {
                            ",", ".", ";", ":", "!", "?",        # Standard sentence punctuation
                            ")", "]", "}",                     # Closing brackets
                            "'s", "'S", "'re", "'RE", "'ve", "'VE", # Common contractions
                            "'m", "'M", "'ll", "'LL", "'d", "'D",
                            "n't", "N'T",
                            "...", "…"                          # Ellipses
                        }
                        if current_word not in no_leading_space_punctuation:
                            sys.stdout.write(' ')
                            self.current_line_length += 1
                    sys.stdout.write(current_word)
                    self.current_line_length += len(current_word)

            if delimiter == '\n':
                sys.stdout.write('\n')
                self.current_line_length = 0
            elif delimiter == ' ':
                # If line is full and a space delimiter arrives, it implies a wrap.
                if self.current_line_length >= self.max_length:
                    sys.stdout.write('\n')
                    self.current_line_length = 0

        sys.stdout.flush()
```

```python
# @title Formatted streaming example
from google.colab import ai

wrapper = LineWrapper()
for chunk in ai.generate_text('Give me a long winded description about the evolution of the Roman Empire.', model_name='google/gemini-2.0-flash', stream=True):
  wrapper.print(chunk)
```
